```python
# Import Dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets,linear_model,metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
#Import Dataset
data_frame =  pd.read_csv('../data/regression1.csv', header=None)

print(data_frame.head())

print(data_frame.describe())
```

```python
# Visualization

# X = Explanatory values/observations
# Y = Corresponding response variable values

#As our data doesn't have headers, we can't use the header names to access columns.
#iloc is a convenient way to access data by index
X = data_frame.iloc[:,0].values.reshape(-1,1)
y = data_frame.loc[:,1]

plt.figure(figsize=(14,10))
plt.title('Explanatory values/observations Vs Corresponding response variable values')
plt.scatter(X,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Create linear regression objest
model = LinearRegression()
model.fit(X,y)
```

```python
# Predict values for X_test data
pred = model.predict(X)
```

```python
# Regression coefficients
print('Regression Coefficients are:',model.coef_)

# Intecept
print('Intercept : ',model.intercept_)


# variance score: 1 means perfect prediction
score = model.score(X, y)
print('Variance score: ', score)

mse = mean_squared_error(y, pred)
# Mean Squared Error
print("Mean squared error: %.2f" % mse )

r2 = r2_score(y, pred)
print("Coefficient of Determination R^2: %.2f" %  r2)
```

```python
# Plot a graph for expected and predicted values
plt.figure(figsize=(14,10))
plt.title('Explanatory values/observations Vs Corresponding response variable values')
plt.scatter(X, y,  color='red')
plt.plot(X, pred, color='blue', linewidth=1)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# splitting X and y into training and testing sets
#Split into 70/30 due to small training data
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30,random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)

pred = model.predict(X_test)
```

```python
# Plot a graph for expected and predicted values
plt.figure(figsize=(14,10))
plt.title('Explanatory values/observations Vs Corresponding response variable values')
plt.scatter(X_test, y_test,  color='red')
plt.plot(X_test, pred, color='blue', linewidth=1)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python

# Regression coefficients
print('Regression Coefficients are:',model.coef_)

# Intecept
print('Intercept : ',model.intercept_)


# variance score: 1 means perfect prediction
score = model.score(X_test, y_test)
print('Variance score: ', score)

mse = mean_squared_error(y_test, pred)
# Mean Squared Error
print("Mean squared error: %.2f" % mse )

r2 = r2_score(y_test, pred)
print("Coefficient of Determination R^2: %.2f" %  r2)
```

## Conclusion / Findings

* From the dataset, we can connote the dependent variable Y (second column) has a linear relationship to the independent variable X (first column).

* Training the model with the 70% dataset and the entire dataset resulted in a higher r^2 value shows that the model would fit over 75% of the data.

*  As the training dataset increases, the r^2 increases as well.
